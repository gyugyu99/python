사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요. 

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
num_list = []
while True:   #무한루프 발동 (숫자를 계속 받기 위함)
    input_num = input("숫자를 입력하세요('done' 입력시 종료됩니다):")
    if input_num == 'done': #input이 done이면
        break #함수 탈출
        
    else:     
        try:  
            int_num = int(input_num) #input 숫자를 정수로 바꿈
            num_list.append(int_num) #num_list에 하나씩 추가
        except ValueError: 
            print(f"잘못된 값:{input_num}. 숫자를 입력해주세요.")
print(num_list)
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time #import 모듈 가져오기

def timer_decorator(func):   #time_decorator 함수 정의
    def wrapper(*args, **kwargs): 
        #내부함수 wrapper 정의 *args, **kwargs 가변매개변수 func대신 실행되고
        #매개변수 갯수, 이름에 상관없이 모두 그대로 받아 처리할 수 있게 해줌
        start_time = time.time() #time 모듈의 time함수를 불러 시작 시간에 적용
        result = func(*args, **kwargs) #원래 함수를 실행한 결과값
        end_time = time.time() #함수 실행 직후 시간 기록
        print(f"함수 실행 시간: {end_time - start_time:.2f}초")
        #함수 실행시간 출력
        return result #wrapper함수는 원래 함수의 반환앖을 그대로 반환
    return wrapper #timer_decorator 함수는 wrapper함수 자체 반환
    #wrapper 함수가 func함수를 대체함

@timer_decorator #데코레이터 실행
def example_function(): #example_function이 timer_decorator 처럼 작동함
    # example_function은 데코레이터가 반환한 wrapper함수를 가리킴
    sum = 0 
    for i in range(1000000):
        sum += i
    return sum

example_function() #이 함수를 호출하면 실제로는 timer_decorator 냅의 wrapper함수가 실행됨.
````
